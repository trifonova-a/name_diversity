```python
data_dir = '../../Downloads/namesbystate'
data_files = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('TXT')]
dataframes = []
for f in data_files:
    dataframes.append(pd.read_csv(f, names=['state', 'sex', 'year', 'name', 'count']))
names_df = pd.concat(dataframes)
```

```python
names_df.head()
```

```python
h = plt.figure(figsize=(10, 7))
plt.plot(names_df.loc[names_df.state == 'NY'].groupby('year')['name'].count())
plt.plot(names_df.loc[names_df.state == 'NJ'].groupby('year')['name'].count())
plt.plot(names_df.loc[names_df.state == 'MS'].groupby('year')['name'].count())
plt.plot(names_df.loc[names_df.state == 'CA'].groupby('year')['name'].count())
plt.plot(names_df.loc[names_df.state == 'TX'].groupby('year')['name'].count())
plt.xlabel('year')
plt.ylabel('Count of different given baby names')
plt.legend(['NY', 'NJ', 'MS', 'CA', 'TX'])
plt.grid()
```

```python
names_grouped_df = names_df.groupby(['name', 'year'])['count'].sum().reset_index()
```

```python
h = plt.figure(figsize=(10, 7))
plt.plot(names_grouped_df.loc[names_grouped_df.name == 'Kobe']['year'], 
         names_grouped_df.loc[names_grouped_df.name == 'Kobe']['count'])
```
